```python
import requests
import pandas as pd
import time

def okx_candles(instId="BTC-USDT", bar="1D", limit=1000):
    url = "https://www.okx.com/api/v5/market/candles"
    params = {"instId": instId, "bar": bar, "limit": str(limit)}
    r = requests.get(url, params=params, timeout=20)
    data = r.json()

    if data.get("code") != "0":
        raise ValueError(f"OKX API error: {data}")

    cols = ["ts","o","h","l","c","vol","volCcy","volCcyQuote","confirm"]
    df = pd.DataFrame(data["data"], columns=cols)

    df["date"] = pd.to_datetime(df["ts"].astype(float), unit="ms")
    df["close"] = df["c"].astype(float)
    df = df.sort_values("date").set_index("date")[["close"]]
    return df

btc = okx_candles("BTC-USDT", bar="1D", limit=1500)
eth = okx_candles("ETH-USDT", bar="1D", limit=1500)

btc["ret"] = btc["close"].pct_change()
eth["ret"] = eth["close"].pct_change()

btc.dropna(inplace=True)
eth.dropna(inplace=True)

btc.tail(), eth.tail()
```

```python
def hist_var_cvar(x: pd.Series, alpha=0.05):
    x = x.dropna()
    var = x.quantile(alpha)
    cvar = x[x <= var].mean()
    return var, cvar

for a in [0.05, 0.01]:
    var_btc, cvar_btc = hist_var_cvar(rets["BTC"], a)
    var_eth, cvar_eth = hist_var_cvar(rets["ETH"], a)
    print(f"Hist {int((1-a)*100)}%  BTC VaR={var_btc:.4%}, CVaR={cvar_btc:.4%} | ETH VaR={var_eth:.4%}, CVaR={cvar_eth:.4%}")
```

```python
import math
from statistics import NormalDist

def param_var_cvar_normal(x: pd.Series, alpha=0.05):
    x = x.dropna()
    mu = x.mean()
    sigma = x.std()
    z = NormalDist().inv_cdf(alpha)  # alpha 分位的 z 值（负数）
    var = mu + sigma * z
    # 正态 CVaR 公式：mu - sigma * phi(z)/alpha
    phi = (1 / math.sqrt(2*math.pi)) * math.exp(-0.5 * z*z)
    cvar = mu - sigma * phi / alpha
    return var, cvar

for a in [0.05, 0.01]:
    var_btc, cvar_btc = param_var_cvar_normal(rets["BTC"], a)
    var_eth, cvar_eth = param_var_cvar_normal(rets["ETH"], a)
    print(f"Normal {int((1-a)*100)}% BTC VaR={var_btc:.4%}, CVaR={cvar_btc:.4%} | ETH VaR={var_eth:.4%}, CVaR={cvar_eth:.4%}")
```

```python
def event_stats(x: pd.Series):
    x = x.dropna()
    cum = (1 + x).prod() - 1
    mdd = ( (1+x).cumprod() / (1+x).cumprod().cummax() - 1 ).min()
    worst = x.min()
    return pd.Series({
        "cum_return": cum,
        "max_drawdown": mdd,
        "worst_day": worst,
        "vol_30d": x.rolling(30).std().iloc[-1] * np.sqrt(365) if len(x) >= 30 else np.nan
    })

def run_event(name, start, end):
    w = rets.loc[start:end]
    out = pd.concat([event_stats(w["BTC"]), event_stats(w["ETH"])], axis=1)
    out.columns = ["BTC", "ETH"]
    print(f"\n=== {name}: {start} to {end} ===")
    display(out)

# 你可以先跑这些（日期不一定完全覆盖，覆盖不到也没事）
run_event("Mar-2020 Crash (312 window)", "2020-03-01", "2020-04-15")
run_event("FTX Collapse window", "2022-11-01", "2022-12-15")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

prices = pd.concat(
    [btc["close"].rename("BTC"), eth["close"].rename("ETH")],
    axis=1
).dropna()

rets = prices.pct_change().dropna()

# 组合：60/40（你也可以改）
w = np.array([0.6, 0.4])
port_ret = pd.Series(rets[["BTC","ETH"]].values @ w, index=rets.index, name="PORT")
```

```python
def rolling_var_backtest(x: pd.Series, window=120, alpha=0.05):
    x = x.dropna()
    var = x.rolling(window).quantile(alpha).shift(1)  # 用过去 window 估计，预测下一天
    breach = (x < var).astype(int)
    out = pd.DataFrame({
        "ret": x,
        "VaR": var,
        "breach": breach
    }).dropna()
    out["breach_rate"] = out["breach"].expanding().mean()
    return out

bt_btc  = rolling_var_backtest(rets["BTC"], window=120, alpha=0.05)
bt_eth  = rolling_var_backtest(rets["ETH"], window=120, alpha=0.05)
bt_port = rolling_var_backtest(port_ret,  window=120, alpha=0.05)

def summarize_bt(bt: pd.DataFrame, alpha=0.05):
    n = len(bt)
    b = int(bt["breach"].sum())
    rate = b / n if n else np.nan
    # 理论期望 breach rate ~ alpha
    return pd.Series({
        "n_days": n,
        "breaches": b,
        "breach_rate": rate,
        "expected": alpha
    })

summary = pd.concat([
    summarize_bt(bt_btc).rename("BTC"),
    summarize_bt(bt_eth).rename("ETH"),
    summarize_bt(bt_port).rename("PORT"),
], axis=1)

summary
```

```python
def plot_var(bt: pd.DataFrame, title="Rolling VaR Backtest"):
    plt.figure()
    plt.plot(bt.index, bt["ret"], label="Return")
    plt.plot(bt.index, bt["VaR"], label="VaR(95%)")
    # breach 点
    breach_idx = bt.index[bt["breach"] == 1]
    plt.scatter(breach_idx, bt.loc[breach_idx, "ret"], label="Breach", s=12)
    plt.title(title)
    plt.legend()
    plt.show()

plot_var(bt_btc,  "BTC: Rolling Historical VaR(95%) Backtest (window=120D)")
plot_var(bt_port, "Portfolio(60/40): Rolling Historical VaR(95%) Backtest (window=120D)")
```

```python
worst = port_ret.nsmallest(10).to_frame("PORT_ret")
worst["BTC_ret"] = rets.loc[worst.index, "BTC"]
worst["ETH_ret"] = rets.loc[worst.index, "ETH"]
worst
```

```python
def window_report(start, end):
    w = rets.loc[start:end]
    pr = port_ret.loc[start:end]
    equity = (1 + pr).cumprod()
    dd = equity / equity.cummax() - 1
    return pd.DataFrame({
        "PORT_cum": equity,
        "PORT_dd": dd,
        "BTC_ret": w["BTC"],
        "ETH_ret": w["ETH"],
        "PORT_ret": pr
    })

# 取组合最差那天作为中心点
center = worst.index[0]
start = (center - pd.Timedelta(days=10)).strftime("%Y-%m-%d")
end   = (center + pd.Timedelta(days=10)).strftime("%Y-%m-%d")

report = window_report(start, end)
report.head(), report.tail()
```

```python
plt.figure()
plt.plot(report.index, report["PORT_dd"])
plt.title(f"Portfolio Drawdown Around Worst Day ({start} to {end})")
plt.show()
```

```python
# 1) 最差一天（组合）
worst_day_ret = report["PORT_ret"].min()
worst_day_date = report["PORT_ret"].idxmin()

# 2) 窗口内最大回撤（越负越大）
max_dd = report["PORT_dd"].min()
max_dd_date = report["PORT_dd"].idxmin()

# 3) 窗口累计收益（窗口起点到终点的累计变化）
window_cum_return = report["PORT_cum"].iloc[-1] / report["PORT_cum"].iloc[0] - 1

worst_day_ret, worst_day_date, max_dd, max_dd_date, window_cum_return
```
